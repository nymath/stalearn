```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from stalearn.supervised.tree import DecisionTreeClassifier as myClassifier
from stalearn.supervised.tree import visualize_tree
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data # + np.random.normal(0, 1, size=iris.data.shape)
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1234)
```

# Compare the base model

```python
model = myClassifier(random_state=42)
clf = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
clf.fit(X_train, y_train)
```

```python
# 使用 plot_tree 来展示决策树结构
plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True);
```

```python
visualize_tree(model.tree_)
```

```python
clf.cost_complexity_pruning_path(X_train, y_train)
```

```python
model.cost_complexity_pruning_path(X_train, y_train)
```

```python
clf.feature_importances_
```

```python
model.feature_importance
```

# Pruning Trees 

```python
model = myClassifier(random_state=42, ccp_alpha=0.05)
clf = DecisionTreeClassifier(random_state=42, ccp_alpha=0.05)
model.fit(X_train, y_train)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(8, 5))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True);
```

```python
visualize_tree(model.tree_)
```

# HyperParameters

```python
model = myClassifier(random_state=42, max_depth=3, min_samples_split=4, min_samples_leaf=2)
clf = DecisionTreeClassifier(random_state=42, max_depth=3, min_samples_split=4, min_samples_leaf=2)
model.fit(X_train, y_train)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(8, 5))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True);
```

```python
visualize_tree(model.tree_)
```
